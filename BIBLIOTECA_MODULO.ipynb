```python
import matplotlib.pyplot as plt

#Dados
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

#Cria um gráfico de linha
#Tem tambem o plt.bar que ai o gráfico fica em barra
plt.bar(x, y)

#Adicionar rótulos aos eixos
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

#Adicionar um título ao gráfico
plt.title('Exemplo de Gráfico de barra')

#Mostrar o gráfico
plt.show()
```

```python
#Vamos tentar deixar o código mais bonito:
import matplotlib.pyplot as plt

#Quais vão ser os dados?
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio']
vendas = [120, 90, 150, 80, 200]

#Criando o gráfico:
# figsize aceita tupla (largura, altura)
plt.figure(figsize=(10, 6)) #Define um tamanho maior para o gráfico

#Qual vai ser o estilo do gráfico?
# ATENÇÃO: 'off white' não é uma cor padrão do Matplotlib e pode falhar.
# Sugiro usar 'lightgray' ou um código hexadecimal, mas mantive sua intenção.
plt.bar(meses, vendas, color = 'lightpink', edgecolor = 'lightgray', alpha=0.7)

#Adicionando títulos e rótulos (ESSENCIAL PARA TER UM GRÁFICO CLARO)
plt.title('Vendas Mensais por Período', fontsize=16, fontweight='bold')
plt.xlabel('Mês', fontsize=12) #Esse é o rótulo do eixo x
plt.ylabel('Valor de venda R$:', fontsize=12) #Esse é o rótulo do eixo y

#Adicionar linhas de Grade (Pra faciliar a leitura dos valores)
plt.grid(axis='y', linestyle='--', alpha=0.5)

#Pra adicionar o  rótulo de dados (para saber o valor exato )
#Loop para calcular o valor exato
for i, v in enumerate(vendas):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', fontweight='bold')

#Pra melhorar a apresentação dos rotulos no eixo x
plt.xticks(rotation=45, ha='right') #Gira os rótulos dos meses para evitar sobreposição

#Ajustando o layout para garantir que tudo caiba certinho:
plt.tight_layout()

#Para mostrar o Gráfico:
plt.show()
```
